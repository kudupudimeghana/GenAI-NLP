```python
import random
human_texts=[
    "I had a great day at college and enjoyed learning new things.",
    "Cybersecurity is becoming very important in todays's world."
]
ai_texts=[
    "The rapid evolution of artificial intelligence is transforming industries globally.",
    "Generative AI models can produce human-like text with remarkable coherence."
]
all_texts=human_texts+ai_texts
random.shuffle(all_texts)
for text in all_texts:
  print("\nText:",text)
  guess=input("Is this AI or Human?").lower()
  if text in ai_texts  and guess == "ai":
    print("Correct!")
  elif text in human_texts and guess == "human":
    print("Correct")
  else:
    print("Oops!wrong")
```

**Basic AI chatbot program**

```python
def chatbot(user_input):
  responses = {
      "hello": "Hi",
      "what is ai":"AI is intelligence shown by machines.",
      "bye":"Good bye! feel free to ask if you have any doubts i am always here to help youuu"
  }
  return responses.get(user_input.lower(),"I don't understand")
while True:
  user_input=input("You:")
  if user_input.lower() == "exit":
    break
  print("Bot:",chatbot(user_input))
```

```python
def restaurant_chatbot(user_input):
    user_input = user_input.lower()

    if "hello" in user_input or "hi" in user_input:
        return "Hello! Welcome to Food Corner Restaurant. How can I help you today?"

    elif "menu" in user_input:
        return "Our menu includes Pizza, Burger, Pasta, Sandwich, Coffee, and Juice."

    elif "pizza" in user_input:
        return "We have Veg Pizza, Cheese Pizza, and Chicken Pizza."

    elif "burger" in user_input:
        return "We have Veg Burger and Chicken Burger."

    elif "pasta" in user_input:
        return "We serve White Sauce Pasta and Red Sauce Pasta."

    elif "coffee" in user_input:
        return "We have Hot Coffee and Cold Coffee."

    elif "juice" in user_input:
        return "We have Orange Juice, Mango Juice, and Apple Juice."

    elif "order" in user_input:
        return "Sure! Please tell me what you would like to order."

    elif "bye" in user_input:
        return "Thank you for visiting Food Corner Restaurant. Have a nice day!"

    else:
        return "Sorry, I didn't understand. Please ask about the menu or place an order."


while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Session ended. Thank you!")
        break

    print("Bot:", restaurant_chatbot(user_input))
```

AI story Genereator using Prompts

```python
def generate_story(prompt, genre, length):
    intro = f"Once upon a time in a {genre.lower()} world, "
    middle = f"there was a strange event related to '{prompt}'. "

    if length == "short":
        body = (
            "A brave character decided to face the challenge. "
            "After a few difficulties, everything ended happily."
        )
    elif length == "medium":
        body = (
            "A brave character decided to face the challenge. "
            "They travelled through many obstacles and discovered hidden secrets. "
            "With courage and kindness, they finally solved the mystery."
        )
    else:
        body = (
            "A brave character decided to face the challenge. "
            "The journey was long and full of mystery, danger, and emotional moments. "
            "Along the way, they met new friends, faced enemies, and learned important life lessons. "
            "In the end, they overcame the challenge and changed their world forever."
        )

    ending = " And so, the story became a legend."
    return intro + middle + body + ending


# User Input
prompt = input("Enter your story prompt: ")
genre = input("Enter genre (Fantasy / Horror / Romance / Sci-Fi / Adventure): ")
length = input("Enter story length (short / medium / long): ")

# Generate Story
story = generate_story(prompt, genre, length)

print("\n✨ Generated Story:\n")
print(story)
```
